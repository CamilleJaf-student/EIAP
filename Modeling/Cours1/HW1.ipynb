From these two values, the total duration of daylight is calculated and expressed in seconds (daylight_s).

Together, these variables allow for the analysis of seasonal changes in day length and solar cycles across different geographic locations. 

The files therefore serve as a comprehensive resource for examining sunlight exposure patterns and their variation throughout the year in each city.

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load CSV files
quito = pd.read_csv("../EIAP/Modeling/Cours1/quito_sun_2023.csv")
rome = pd.read_csv("../EIAP/Modeling/Cours1/rome_sun_2023.csv")
copenhagen = pd.read_csv("../EIAP/Modeling/Cours1/copenhagen_sun_2023.csv")
tromso = pd.read_csv("../EIAP/Modeling/Cours1/tromso_sun_2023.csv")

# Convert 'date' column to datetime
quito['date'] = pd.to_datetime(quito['date']).dt.normalize()
rome['date'] = pd.to_datetime(rome['date']).dt.normalize()
copenhagen['date'] = pd.to_datetime(copenhagen['date']).dt.normalize()
tromso['date'] = pd.to_datetime(tromso['date']).dt.normalize()

# Plot daylight_h
plt.figure(figsize=(12,6))
plt.plot(quito['date'], quito['daylight_h'], label="Quito")
plt.plot(rome['date'], rome['daylight_h'], label="Rome")
plt.plot(copenhagen['date'], copenhagen['daylight_h'], label="Copenhagen")
plt.plot(tromso['date'], tromso['daylight_h'], label="Tromsø")

# Format x-axis as months
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b'))

plt.xlabel("Month")
plt.ylabel("Daylight hours")
plt.title("Daylight duration in 2023 for three cities")
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

When adding Tromsø, located above the Arctic Circle, who experiences several weeks of continuous daylight in summer (the “midnight sun”), with 24 hours of illumination, and conversely enters the polar night in winter, where the Sun does not rise at all and daylight hours fall to zero. 

This generates a characteristic plateau at the top and bottom of the curve, unlike the smooth oscillations observed at lower latitudes.

These differences are explained by the axial tilt of the Earth (23.5°) and the latitude of each city. 

Near the equator, the tilt has little effect on day length, while at increasing latitudes the tilt produces progressively larger seasonal contrasts. 

In summary, the four cities illustrate the continuum of daylight dynamics across latitudes: from constant near the equator (Quito), to moderate oscillations (Rome), to strong seasonal swings (Copenhagen), and finally to the extreme alternations of continuous day and night at polar latitudes (Tromsø).
